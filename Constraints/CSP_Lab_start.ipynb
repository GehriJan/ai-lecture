## Variablen und Wertebereiche festlegen

```python
from constraint import *

# CSP initialisieren
csp = Problem()

#Variablen definieren und Wertebereiche festlegen
csp.addVariable("A", [i for i in range(0,25+1)])
csp.addVariable("B", [i for i in range(0,100+1)])
csp.addVariable("C", [i for i in range(0,20+1)])
csp.addVariable("D", [i for i in range(0,10+1)])

# ... Sie sind dran

# am Schluss kurz alle Kombinationen ausgeben (zur Kontrolle)

csp.getSolutions()
```

## Lösungsmenge für Volumen und Gewicht

```python
# Volumenlimit 2500

def testlimit(a,b,c,d):
    return (100*a + 50*b + 70*c + 25*d < 2501)
    
    
csp.addConstraint(testlimit, ["A","B","C","D"])
    
# Gewichtslimit 2800

def testlimit_42(a,b,c,d):
    return (120*a + 175*b + 140*c + 100*d < 2801)
    
    
csp.addConstraint(testlimit_42, ["A","B","C","D"])
    
```

```python
csp.getSolutions()
```
